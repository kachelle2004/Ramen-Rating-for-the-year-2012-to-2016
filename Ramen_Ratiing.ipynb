```python
df['Top Ten']= df['Top Ten'].replace("\n",np.nan)
```

```python
df[df['Top Ten'].notna()]
```

```python
# 1. Filter the dataframe
top_ten_df = df[df['Top Ten'].notna()].copy()

# 2. Extract Year and Rank into new columns
# This splits "2016 #10" into 2016 and 10
top_ten_df[['Year', 'Rank']] = top_ten_df['Top Ten'].str.extract(r'(\d+)\s#(\d+)').astype(int)

# 3. Sort by Year (ascending) and then Rank (ascending)
sorted_df = top_ten_df.sort_values(by=['Year', 'Rank'], ascending=[True, True])

# 4. Optional: Remove the temporary helper columns
sorted_df = sorted_df.drop(columns=['Year', 'Rank'])

```

```python
# plot a basic vertical bar graph between the brand and top ten
plt.bar( sorted_df['Top Ten'],sorted_df['Brand'], color='skyblue')

# Add labels and a title
plt.xlabel('Top Ten Rankings')
plt.ylabel('Brands')
plt.title('Brands being awarded Top Ten Rankings')

# Rotate x-axis labels 45 degrees
plt.xticks(rotation=45, ha='right')  # ha='right' makes the labels align nicely

# Display the plot
plt.show()
```

```python


# Count how many times each brand appears
brand_counts = sorted_df['Brand'].value_counts()  # Series: index = Brand, value = count

# Plot
plt.bar(brand_counts.index, brand_counts.values, color='skyblue')

plt.xlabel('Brands')
plt.ylabel('Count in Top Ten')
plt.title('Number of Times Each Brand Appeared in Top Ten')

plt.xticks(rotation=45, ha='right')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Count how many times each brand appears
country_counts = sorted_df['Country'].value_counts()  

# Plot
plt.bar(country_counts.index, country_counts.values, color='skyblue')

plt.xlabel('Country')
plt.ylabel('Count in Top Ten')
plt.title('Number of Times Each Country Appeared in Top Ten')

plt.xticks(rotation=45, ha='right')

plt.show()
```
